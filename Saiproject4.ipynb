# Predictive Modeling:

## Research Question:
Can we predict the Clean Alternative Fuel Vehicle (CAFV) eligibility of an electric vehicle based on its attributes?

## Analysis
* Data Preprocessing:

  The method dubbed "preprocess_data" takes on several critical tasks, namely, grappling with the void left behind by missing values, executing categorical variable encoding, and, all in all, getting all the ducks in a row for model training.
* Model Training:

  In the task of constructing a model capable of accurately predicting future observations, the train_model method plays a pivotal role. This imperative function facilitates the training process by utilizing a RandomForestClassifier algorithm on meticulously preprocessed data, eventually churning out a well-rounded and effective predictive model.
* Results:

  * The measurement of the accuracy score is a fundamental validator in determining how well a model can predict CAFV eligibility.

  * An observation of model performance across different classes is facilitated via an in-depth depiction of true positive, false positive, true negative and false negative predictions through the confusion matrix.

Upon rigorous analysis of the available data, it can be inferred that by deploying provided attributes, one might potentially prognosticate whether an electric vehicle possesses the desirable qualifications necessary to possess a Clean Alternative Fuel Vehicle status.


```python
positive_class_label = 1
predictive_modeling = PredictiveModeling(df)
```

```python
predictive_modeling.preprocess_data()
```

```python
accuracy, confusion_matrix, probas, y_test = predictive_modeling.train_model('Clean Alternative Fuel Vehicle (CAFV) Eligibility', positive_class_label)
print(f'Accuracy: {accuracy}')
```

```python
print(f'Confusion Matrix:\n{confusion_matrix}')
```

```python
predictive_modeling.plot_confusion_matrix(confusion_matrix)
```

```python
predictive_modeling.plot_roc_curve(y_test, probas)
```
